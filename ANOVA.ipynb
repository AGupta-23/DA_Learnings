````python
import seaborn as sns #titanic dataset preloaded
import pandas as pd #dataframe manipulation
from scipy.stats import f_oneway  # statistical testing - scipy # anova test - foneway
````

````python
df= sns.load_dataset('titanic')
df
````

````python
df = df[['age', 'pclass']].dropna()
````

````python
#compare - we have 2 genders and ages and passenger class

df['pclass'].unique()

#is there any significant relation between ages and passenger class
# 1 is categorical and 1 is quantitative data
````

````python
class_1 = df[df['pclass'] == 1]['age']
class_2 = df[df['pclass'] == 2]['age']
class_3 = df[df['pclass'] == 3]['age']
````

````python
class_1
````

````python
# Null Hypothesis - no connection between ages and passenger class
# Alternate Hypothesis - Yes connection

f_stats, p_value = f_oneway(class_1, class_2, class_3)
f_stats
````

````python
p_value
````

````python
alpha = 0.05
````

````python
if p_value< alpha:
    print("We will reject the null hypo and there is significant difference in atleast 1 passenger class")
else: 
    print("Accept the null hypothesis - there is significant difference")
````

````python
## This proved that all 3 passenger classes had age distributions in passengers
````

# ANOVA (Analysis of Variance)

## Definition
ANOVA (Analysis of Variance) is a statistical test used to compare the means of **three or more groups** and determine whether at least one group mean is significantly different from the others.

---

## When to Use ANOVA
Use ANOVA when:
- One variable is **categorical** (groups/categories).
- One variable is **quantitative/numerical**.
- There are **3 or more groups**.

### Examples
- Age vs Passenger Class (1st, 2nd, 3rd)
- Salary vs Department
- Marks vs Teaching Method
- Blood Pressure vs Different Medicines

---

## Why Not Multiple t-tests?
If we perform many t-tests, the probability of making a Type I error increases.

ANOVA performs a **single test** to compare all group means simultaneously.

---

## Hypotheses

### Null Hypothesis (H₀)
All population means are equal.

μ₁ = μ₂ = μ₃ = ... = μₖ

### Alternative Hypothesis (H₁)
At least one group mean is different.

---

## Basic Idea
ANOVA compares:

1. **Between-group variation**
   - Difference between group means.

2. **Within-group variation**
   - Variation inside each group.

If the between-group variation is much larger than the within-group variation, the means are likely different.

---

## F Statistic

F = (Between-group variance) / (Within-group variance)

- Large F-value → Means are significantly different.
- Small F-value → Means are similar.

---

## Decision Rule

If:

p-value < α

Reject H₀ → At least one group mean is different.

If:

p-value ≥ α

Fail to reject H₀ → No significant difference between group means.

---

# In This Example

Categorical Variable:
- `pclass` (1st, 2nd, 3rd class)

Numerical Variable:
- `age`

Question:
**Do passengers of different classes have different average ages?**

Results:

- F-statistic = 57.44
- p-value = 7.49 × 10⁻²⁴
- α = 0.05

Since:

7.49 × 10⁻²⁴ < 0.05

Reject H₀.

**Conclusion:** At least one passenger class has a significantly different average age.
