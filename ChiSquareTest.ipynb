# Chi-Square Test (χ² Test of Independence)

## Definition
The Chi-Square Test of Independence is a statistical test used to determine whether there is a significant relationship between two categorical variables.

It checks whether the variables are:
- Independent (no relationship)
- Dependent (some relationship exists)

---

## When to Use Chi-Square Test
Use this test when:
1. Both variables are categorical.
2. Data is in the form of frequencies or counts.
3. You want to check association between categories.

### Examples
- Gender vs Survival
- Education Level vs Placement
- Smoking Status vs Disease
- Department vs Promotion

---

## Hypotheses

### Null Hypothesis (H₀)
The two categorical variables are independent.
(No relationship exists.)

### Alternative Hypothesis (H₁)
The two categorical variables are dependent.
(A relationship exists.)

---

## Basic Idea
Chi-Square compares:

- Observed Frequencies (Actual counts in the data)
- Expected Frequencies (Counts expected if there were no relationship)

If the observed and expected frequencies differ greatly, there is evidence of a relationship between the variables.

---

## Formula

χ² = Σ [(O − E)² / E]

Where:
- O = Observed Frequency
- E = Expected Frequency

---

## Degrees of Freedom

df = (Rows − 1) × (Columns − 1)

For a 2 × 2 table:

df = (2 − 1)(2 − 1) = 1

---

## Decision Rule

If:

p-value < α

Reject H₀ → Significant relationship exists.

If:

p-value ≥ α

Fail to reject H₀ → No significant relationship exists.
