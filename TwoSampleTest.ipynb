````python
import numpy as np
from scipy import stats  #stats offer two sample test directly
````

````python
group_A= [85,88,90,92,87,85,89,91,86,88]
group_B= [82,84,80,83,81,79,78,85,84,83]
````

````python
t_stats , p_value = stats.ttest_ind(group_A, group_B, equal_var= False)

print(t_stats)
print(p_value)
````

````python
alpha= 0.05

if p_value < alpha:
    print("I will reject the null hypothesis")
else: 
    print("I will accept the null hypo")
````

# Two Sample T-Test

## What is it?

A Two Sample T-Test checks whether the means of **two independent groups** are significantly different.

Question it answers:

> "Is the difference between the two sample averages real, or did it happen by random chance?"

---

## Example

Suppose:

- Group A = Students taught using Method A
- Group B = Students taught using Method B

Scores:

```python
group_A = [85,88,90,92,87,85,89,91,86,88]
group_B = [82,84,80,83,81,79,78,85,84,83]
```

Average scores:

```text
Group A ≈ 88.1
Group B ≈ 81.9
Difference ≈ 6.2 marks
```

Question:

> Did Method A actually perform better, or is this difference just luck?

The Two Sample T-Test answers this.

---

# Hypotheses

### Null Hypothesis (H₀)

```text
μA = μB
```

Both population means are equal.

### Alternative Hypothesis (H₁)

```text
μA ≠ μB
```

The means are different.

---

# Formula

\[
t=\frac{\bar{x}_1-\bar{x}_2}{SE}
\]

where,

- \(\bar{x}_1\) = mean of Group A
- \(\bar{x}_2\) = mean of Group B
- SE = standard error of the difference

---

# Your Result

```text
t = 5.83
```

This is a very large t-value.

Meaning:

> The difference between the two sample means is much larger than what random chance usually produces.

So the p-value becomes very small.

Since:

```text
p < 0.05
```

Decision:

```text
Reject H₀
```

Conclusion:

> The two groups have significantly different means.

---

# Code (Short Explanation)

### Import

```python
from scipy import stats
```

Imports statistical functions.

---

### Data

```python
group_A = [...]
group_B = [...]
```

Creates the two samples.

---

### T-Test

```python
t_stats, p_value = stats.ttest_ind(
                    group_A,
                    group_B,
                    equal_var=False)
```

- `ttest_ind()` → Independent Two Sample T-Test
- `t_stats` → calculated t-value
- `p_value` → probability of getting this result by chance
- `equal_var=False` → uses Welch's T-Test (safer when variances may differ)

---

### Significance Level

```python
alpha = 0.05
```

5% significance level.

---

### Decision

```python
if p_value < alpha:
    print("Reject H0")
else:
    print("Accept H0")
```

Rule:

```text
p < α  → Reject H₀
p ≥ α → Fail to Reject H₀
```

Since your p-value is very small:

```text
Reject H₀
```

The average scores of Group A and Group B are significantly different.
