````python
import numpy as np
from scipy.stats import norm  #norm- normal distribution
````

````python
sample = [172,174, 168, 169, 171,173, 175, 170, 169, 172]
population_mean= 170
population_std=3
samp_mean= np.mean(sample)
n= len(sample)
````

````python
z_score = (samp_mean - population_mean) / (population_std / np.sqrt(n))
z_score
````

````python
alpha = 0.05
````

````python
p_value= 2* (1-norm.cdf(abs(z_score)))
p_value
````

````python
# if p value is < alpha -- reject null hypo
# if p value is > alpha -- accept null hypo

if p_value < alpha:
    print("I will reject the null hypothesis")
else:
    print("I will accept the null hypothesis")
````

# Z-Test

## What is a Z-Test?
A Z-test checks whether the difference between a sample mean and a population mean is **statistically significant**.

### Use Z-Test when:
- Population standard deviation (σ) is known.
- Sample size is large (n ≥ 30) OR data is approximately normal.

---

## Why do we use it?
A sample mean may differ from the population mean simply due to random chance.

The Z-test tells us:

**"Is the difference large enough to conclude that the population mean has actually changed?"**

---

# Why do we use P-values in Coding?

Finding critical values manually is easy only for common α values like:

- α = 0.05
- α = 0.01
- α = 0.10

In coding, α can be anything:

- 0.03
- 0.08
- 0.123
- etc.

Instead of looking up Z tables, we calculate the exact probability using the p-value.

---

## Meaning of P-value

P-value = probability of getting a result this extreme (or more extreme) assuming H₀ is true.

For your code:

```python
p_value = 2 * (1 - norm.cdf(abs(z_score)))
```

- `norm.cdf()` gives left-side probability.
- `1 - norm.cdf()` gives right-tail probability.
- `2 *` is used because it is a two-tailed test.

Result:

```python
p_value = 0.1706
```

Decision rule:

```python
if p_value < alpha:
    Reject H0
else:
    Fail to Reject H0
```

Since,

```python
0.1706 > 0.05
```

### Conclusion:
Fail to reject H₀.

---
